
###  7. Size
_**Clean & Convert 'Size' Column to Numeric (MB)**_
The `Size` column contains string values like `'12M'`, `'800k'`, and `'Varies with device'`.

To prepare it for analysis:
- We first apply a custom function to:
  - Remove characters like `'M'` and `'k'`
  - Convert values to proper numeric format (MB)
  - Handle unknown formats by returning `None`
- _**Then we fill the missing values with the **mean size** to ensure the column is fully numeric and ready for modeling.**_


<a id="ok00"></a>
<h2 style="display:inline-block;
           background-color:#1c0033;
           color:#ff66cc;

*Note*

>_"This step technically belongs to feature engineering, but it is added here to ensure proper handling of null values, which couldnâ€™t be filled without thisÂ adjustment."_ \
[Check using this](#ok11)

```python
import numpy as np
# convert the size column to numeric by multiplying the values with 1024 if it has 'k' in it and 1024*1024 if it has 'M' in it
# this function will convert the size column to numeric
def convert_size(size):
        # add function details here
    '''
    This function will convert the size column to numeric by multiplying the values with 1024 if it has 'k' in it and 1024*1024 if it has 'M' in it
    '''
    
    if isinstance(size, str):
        size = size.replace(',', '').strip()  # Remove commas and spaces
        if 'k' in size:
            return float(size.replace('k', '')) * 1024  # Convert KB to Bytes
        elif 'M' in size:
            return float(size.replace('M', '')) * 1024 * 1024  # Convert MB to Bytes
        elif 'G' in size:
            return float(size.replace('G', '')) * 1024 * 1024 * 1024  # If GB exists
        elif 'Varies with device' in size or size == '':
            return np.nan
    return np.nan  # For non-string or unexpected formats

df['Size'] = df['Size'].apply(convert_size)
```

```python
# Fill missing values with mean size
df['Size'].fillna(df['Size'].mean(), inplace=True)
```


###  10. Installs
 _**Clean & Convert 'Installs' Column to Numeric**_
- The `Installs` column contains string values like `'1,000+'`, `'500,000+'`, etc.

To prepare this column:
- We first **remove '+' and ','** symbols to clean the data.
- Then we **convert** the column to a numeric format using `pd.to_numeric()`.
>_Finally, we **fill any missing values** with the **median**, which handles outliers better than mean._

```python
# Remove '+' and ',' symbols from each value in the 'Installs' column
df['Installs'] = df['Installs'].apply(lambda x: str(x).replace('+', '').replace(',', ''))
# Replace the comas ',' in the installs columns
df['Installs'] = df['Installs'].apply(lambda x: str(x).replace(',', '').replace('+', ''))
# Convert these cleaned values into numbers (integers or floats)
# This will help in doing math or graphs later
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')  # if error, put NaN (missing value)

# : Fill missing with median
df['Installs'].fillna(df['Installs'].median(), inplace=True)
```

- Convert the Size column into numeric bytes using a custom function.



```python
import numpy as np
# convert the size column to numeric by multiplying the values with 1024 if it has 'k' in it and 1024*1024 if it has 'M' in it
# this function will convert the size column to numeric
def convert_size(size):
        # add function details here
    '''
    This function will convert the size column to numeric by multiplying the values with 1024 if it has 'k' in it and 1024*1024 if it has 'M' in it
    '''
    
    if isinstance(size, str):
        size = size.replace(',', '').strip()  # Remove commas and spaces
        if 'k' in size:
            return float(size.replace('k', '')) * 1024  # Convert KB to Bytes
        elif 'M' in size:
            return float(size.replace('M', '')) * 1024 * 1024  # Convert MB to Bytes
        elif 'G' in size:
            return float(size.replace('G', '')) * 1024 * 1024 * 1024  # If GB exists
        elif 'Varies with device' in size or size == '':
            return np.nan
    return np.nan  # For non-string or unexpected formats

df['Size'] = df['Size'].apply(convert_size)
```

```python
# rename the column name 'Size' to 'Size_in_bytes'
df.rename(columns={'Size': 'Size_in_bytes'}, inplace=True)
```

```python
# making a new column called 'Size in Mb' which will have the size in MB
df['Size_in_Mb'] = df['Size_in_bytes'].apply(lambda x: x/(1024*1024))
```

### Observation
- The original Size column had values like `'12M', '500k',` or `'Varies with device'` which were not suitable for numerical analysis.
- We successfully converted all sizes to numeric values in bytes, making them usable for calculations, sorting, and comparisons.
- After converting to bytes, we created a new column `Size_in_Mb` for easier readability and better interpretation.
- Now we can easily analyze app sizes in megabytes `(MB),` which is more practical and user-friendly.

**Detect Outliers using IQR (Interquartile Range**

```python
# Example for 'Price' column
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1

# Find outliers
outliers = df[(df['Price'] < Q1 - 1.5 * IQR) | (df['Price'] > Q3 + 1.5 * IQR)]
print("Total outliers in Price column:", outliers.shape[0])
```

**Before vs After Removing Outliers**

```python
print("Before removing outliers:")
print(df['Price'].describe())

# Remove outliers
df_clean = df[~((df['Price'] < Q1 - 1.5 * IQR) | (df['Price'] > Q3 + 1.5 * IQR))]

print("After removing outliers:")
print(df_clean['Price'].describe())
```
